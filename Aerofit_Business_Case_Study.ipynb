Calculating the IQR for Income column

```python
q1 = np.percentile(af["Income"],25)
q3 = np.percentile(af["Income"],75)
IQR = q3-q1
Upper_band = q3+1.5*(IQR)
Lower_band = q1-1.5*(IQR)
Median = af["Income"].median()

print("Q1=", q1)
print("Q3=", q3)
print("IQR=", IQR)
print("Upper band=", Upper_band)
print("Lower band=", Lower_band)
print("Median", af["Income"].median())
```

All values above the upper band i.e >80581.875 are outliers.

```python
(len(af.loc[af["Income"]>Upper_band])/len(af))*100
```
